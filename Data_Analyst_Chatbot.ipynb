```python
import pandas as pd
from transformers import pipeline
```

```python
# Load your dataset
df = pd.read_csv('/content/laptop_data_cleaned.csv')

# Optionally, perform data cleaning or preprocessing here
# For example:
# df.dropna(inplace=True)  # Remove rows with missing values
```

```python
nlp_pipeline = pipeline("text-generation", model="gpt2")
```

```python
def summarize_data():
    summary = df.describe()
    return summary.to_string()

def analyze_column(column_name):
    if column_name in df.columns:
        column_stats = df[column_name].describe()
        return column_stats.to_string()
    else:
        return f"Column '{column_name}' not found in the dataset."

def generate_insights(prompt):
    generated_text = nlp_pipeline(prompt, max_length=150, num_return_sequences=1, temperature=0.7)
    return generated_text[0]['generated_text']
```

```python
def main():
    print("Welcome! I'm your Data Analyst Chatbot. How can I assist you today?")

    while True:
        user_input = input("You: ")

        if user_input.lower() == 'exit':
            print("Goodbye!")
            break

        elif user_input.lower().startswith('summarize data'):
            summary = summarize_data()
            print(summary)

        elif user_input.lower().startswith('analyze column'):
            _, _, column_name = user_input.partition('analyze column')
            column_name = column_name.strip()
            result = analyze_column(column_name)
            print(result)

        else:
            # Use NLP model for generating insights based on user input
            generated_text = generate_insights(user_input)
            print("Chatbot:", generated_text)

if __name__ == "__main__":
    main()
```
